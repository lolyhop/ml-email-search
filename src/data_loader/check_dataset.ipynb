### Exploratory Data Analysis

```python
dataset: pd.DataFrame = emails_loader.dataset
```

#### Check the emails dates distribution

```python
dataset['email_date'].hist(bins=50, figsize=(10, 5))
plt.title('Email Distribution Over Time')
plt.show()
```

#### Check the emails lengths

```python
lengths = dataset['email_content'].apply(len)
plt.figure(figsize=(10, 5), dpi=150)
lengths.hist(bins=50, alpha=0.7)
plt.axvline(lengths.mean(), color='red', linestyle='--', label=f'Mean: {lengths.mean():.0f}')
plt.axvline(lengths.median(), color='green', linestyle='--', label=f'Median: {lengths.median():.0f}')
plt.title('Email Content Length Distribution')
plt.xlabel('Characters')
plt.legend()
plt.show()
```

One might see that we have outliers in length distribution
